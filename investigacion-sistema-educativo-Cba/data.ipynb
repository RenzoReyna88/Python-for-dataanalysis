```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Procesamiento de datos para escuelas del nivel Primario

```python
# Lectura de archivos para nivel primario
leer_archivo_matriculados_primaria= pd.read_csv("datos-nivel-primario/alumnos_matriculados_primario_1998-2021.csv", sep=';')

leer_archivo_matriculados_primaria_segun_genero= pd.read_csv("datos-nivel-primario/alumnos_matriculados_primario_segun_genero_1998-2021.csv", sep=';')

leer_archivo_repitentes_primaria= pd.read_csv("datos-nivel-primario/alumnos_repitentes_primario_1998-2021.csv", sep=';')

leer_archivo_repitentes_primaria_segun_genero= pd.read_csv("datos-nivel-primario/alumnos_repitentes_primario_segun_genero_1998-2021.csv", sep=';')

leer_archivo_sobredad_primaria= pd.read_csv("datos-nivel-primario/alumnos_con_sobredad_primario_1998-2021.csv", sep=';')
```

```python
# Crear Dataframe para alumnos matriculados en la Prov. de Córdoba
matriculados_primaria= pd.DataFrame(leer_archivo_matriculados_primaria)

matriculados_primaria
```

```python
# Seleccionar fila para Dpto. Totoral
Departamento_Totoral= matriculados_primaria.iloc[24]
```

```python
# Crear Dataframe para alumnos inscritos en escuelas primarias del Dpto. Totoral 
df_matriculados_primaria= pd.DataFrame(Departamento_Totoral)

df_matriculados_primaria_totoral= df_matriculados_primaria.T
```

```python

'''Los datos en las siguientes variables representan la cantidad de alumnos de primaria incritos en el Dpto Totoral 
   para el peridodo comprendido entre el año 1998-2021'''

data_inscritos = [2188, 2193, 2173, 2171, 2253, 2229, 2213, 2199, 2179, 2162, 2104, 2102, 2159, 2151, 2120, 2085, 2078, 2009, 2008, 1987, 2035, 2019, 1984, 2012]
años = [1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

# Estilo de Seaborn
sns.set(style="whitegrid")

# Tamaño de la figura
plt.figure(figsize=(12, 8))

# Crear gráfico de barras
barplot = sns.barplot(x=años, y=data_inscritos, hue=años, palette='viridis', edgecolor='#2c3e50', alpha=0.8, legend=False)

# Agregar etiquetas con las cantidades de alumnos para cada barra
for i, txt in enumerate(data_inscritos):
    barplot.text(i, txt, f'{txt}', ha='center', va='bottom', fontsize=10, color='#333333')

# Configurar etiquetas y título
plt.xlabel('Año', fontsize=14, fontweight='bold')
plt.ylabel('Cantidad de alumnos inscritos', fontsize=14, fontweight='bold')
plt.title('Cantidad de alumnos del Dpto. Totoral inscritos por año - Nivel primario', fontsize=16, fontweight='bold')

# Configurar color de fondo del gráfico
plt.setp(barplot.patches, linewidth=0)

# Ajustar el color de fondo
plt.gca().set_facecolor('#ecf0f1')

# Mostrar el gráfico
plt.tight_layout()
plt.show()

Promedio_por_años= np.mean(data_inscritos)

print(f' Promedio de alumnos de primaria inscritos en el Dpto Totoral entre el periodo 1998 - 2021: {Promedio_por_años}')
```

##### “El gráfico anterior muestra el comportamiento de los alumnos inscritos en las escuelas del Departamento de Totoral. Se puede apreciar que no hubo cambios significativos en las inscripciones hasta el período 2021.”

```python

# Datos proporcionados
cant_inscritos_segun_genero = [1104, 1084, 1118, 1075, 1099, 1074, 1104, 1067, 1136, 1117, 1137, 1092, 1128, 1085, 1121, 1078, 
                                1142, 1037, 1129, 1033, 1095, 1009, 1086, 1016, 1132, 1027, 1122, 1029, 1087, 1033, 1058, 1027, 
                                1046, 1032, 993, 1016, 991, 1017, 986, 1001, 1010, 1025, 998, 1021, 998, 986, 1013, 999]

# Años correspondientes a cada par de valores
years = list(range(1998, 1998 + len(cant_inscritos_segun_genero)//2))

# Asegurar que los datos tienen la longitud correcta
data = cant_inscritos_segun_genero[:len(years) * 2]

# Crear un array de índices para posicionar las barras
ind = np.arange(len(years))

# Configurar el tamaño de la figura
plt.figure(figsize=(15, 8))

# Crear gráfico de barras agrupado con colores mejorados
bar_width = 0.35
plt.bar(ind, data[::2], width=bar_width, label='Varones', color='#6fa4e5', edgecolor='black')
plt.bar(ind + bar_width, data[1::2], width=bar_width, label='Mujeres', color='#f1908e', edgecolor='black')

# Configurar etiquetas y título
plt.xlabel('Año', fontweight='bold', fontsize=12)
plt.ylabel('Cantidad', fontweight='bold', fontsize=12)
plt.title('Cant de Varones y Mujeres inscritos en escuelas del Dpto. Totoral por Año - Nivel primario', fontweight='bold', fontsize=14)

# Agregar cantidades correspondientes a cada barra con etiquetas
for i, valor in enumerate(data[::2]):
    plt.text(ind[i], valor + 10, str(valor), ha='center', va='bottom', fontsize=10, fontweight='bold', color='#3498db')

for i, valor in enumerate(data[1::2]):
    plt.text(ind[i] + bar_width, valor + 10, str(valor), ha='center', va='bottom', fontsize=10, fontweight='bold', color='#e74c3c')

# Configurar posición de los ticks en el eje x
plt.xticks(ind + bar_width / 2, years)

# Mostrar la leyenda
plt.legend(title='Sexo', title_fontsize='12', loc='upper right')

# Mostrar el gráfico
plt.show()

promedio_inscritos_por_año= np.mean(cant_inscritos_segun_genero)

print(f'Promedio de alumnos inscritos en escuelas del Dpto. Totoral entre el periodo 1998- 2021: {promedio_inscritos_por_año}')
```

```python
# Datos correspondientes al porcentaje de alumnos repitentes del nivel primario en el Dpto. Totoral
data = {
    'Año': list(range(1998, 2022)),
    'Totoral': [11.9, 11.6, 13.3, 11.2, 11.1, 6.9, 7.3, 8.5, 7.6, 4.3, 7.6, 6.4, 6.4, 2, 2.2, 2, 1.5, 0.3, 0.9, 1.1, 0.5, 0.1, 0.3, 0.9]
}

nuevo_df = pd.DataFrame(data)

# Configurar el estilo de Seaborn para mejorar la estética
sns.set(style="whitegrid", rc={"grid.linestyle": '--'})

# Configurar el tamaño de la figura y la resolución
fig, ax = plt.subplots(figsize=(12, 9), dpi=80)

# Crear el gráfico de barras
bars = ax.bar(nuevo_df['Año'], nuevo_df['Totoral'], color='#3498db', edgecolor='black', linewidth=1)

# Añadir etiquetas y título
plt.title('Índice de Repitencia por Año', fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('Año', fontsize=14, color='#555555')
plt.ylabel('Porcentaje de Repitencia (%)', fontsize=14, color='#555555')
plt.xticks(fontsize=12, color='#555555')
plt.yticks(fontsize=12, color='#555555')

# Añadir etiquetas en las barras con valores en porcentajes
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{round(yval, 2)}%', ha='center', va='bottom', fontsize=10, color='#555555')

# Añadir una leyenda
plt.legend(['Repitencia'], loc='upper right', fontsize=12)

# Añadir una rejilla para mejor lectura
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Eliminar bordes innecesarios
sns.despine()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

##### “El gráfico anterior evidencia una disminución drástica en el porcentaje de alumnos que repiten el año en el Departamento de Totoral a partir del año 2001. Este descenso no solo transmite un mensaje positivo con respecto a la tasa de repetición, sino que también proporciona un contexto para reflexionar, especialmente considerando que el país estaba atravesando una crisis socioeconómica en ese momento.”
