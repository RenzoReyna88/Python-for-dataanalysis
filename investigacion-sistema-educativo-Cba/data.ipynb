```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
leer_archivo_primaria= pd.read_csv(r"C:\Users\r_ger\Downloads\repitencia-1998-2021-primaria.csv", sep=';')
```

```python
repitencia_primaria= pd.DataFrame(leer_archivo_primaria)

repitencia_primaria.head()
```

```python
nuevo_nombre = ['Departamento'] + [str(año) for año in range(1998, 2022)]  # Crear una lista con los nuevos nombres
repitencia_primaria.columns = nuevo_nombre  # Asignar los nuevos nombres a las columnas

# Eliminar la primera fila
df = repitencia_primaria.drop(0)
```

```python
data = {
    'Año': list(range(1998, 2022)),
    'Totoral': [11.9, 11.6, 13.3, 11.2, 11.1, 6.9, 7.3, 8.5, 7.6, 4.3, 7.6, 6.4, 6.4, 2, 2.2, 2, 1.5, 0.3, 0.9, 1.1, 0.5, 0.1, 0.3, 0.9]
}

nuevo_df = pd.DataFrame(data)

# Configurar el estilo de Seaborn para mejorar la estética
sns.set(style="whitegrid", rc={"grid.linestyle": '--'})

# Configurar el tamaño de la figura y la resolución
fig, ax = plt.subplots(figsize=(12, 9), dpi=80)

# Crear el gráfico de barras
bars = ax.bar(nuevo_df['Año'], nuevo_df['Totoral'], color='#3498db', edgecolor='black')

# Añadir etiquetas y título
plt.title('Índice de repitencia por Año', fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('Año', fontsize=14, color='#555555')
plt.ylabel('Porcentaje de repitencia (%)', fontsize=14, color='#555555')
plt.xticks(fontsize=12, color='#555555')
plt.yticks(fontsize=12, color='#555555')

# Añadir etiquetas en las barras con valores en porcentajes
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{round(yval, 2)}%', ha='center', va='bottom', fontsize=10, color='#555555')

# Añadir una leyenda
plt.legend(['Repitencia'], loc='upper right', fontsize=12)

# Añadir una rejilla para mejor lectura
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Eliminar bordes innecesarios
plt.box(False)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Datos proporcionados
porcentaje = [333, 307, 332, 299, 297, 291, 289, 263, 218, 232, 234, 221, 227, 224, 177, 16, 115, 101, 64, 49, 33, 24, 32, 24]
años = [1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

# Ajustar el porcentaje dividiendo cada valor por 10
porcentaje_ajustado = [p / 10 for p in porcentaje]

# Establecer el estilo de seaborn para una apariencia más moderna
sns.set(style="whitegrid")

# Crear el gráfico con Matplotlib
plt.figure(figsize=(12, 6))
sns.lineplot(x=años, y=porcentaje_ajustado, marker='o', color='#3498db', linewidth=2)

# Añadir etiquetas y título
plt.xlabel('Año', fontsize=14, fontweight='bold')
plt.ylabel('Porcentaje de alumnos con sobreedad (%)', fontsize=14, fontweight='bold')
plt.title('Indice de sobredad por año', fontsize=16, fontweight='bold')

# Añadir etiquetas a los puntos de datos

# Ajustes adicionalesfor i, txt in enumerate(porcentaje_ajustado):
for i, txt in enumerate(porcentaje_ajustado):
    plt.annotate(f'{txt:.1f}%', (años[i], porcentaje_ajustado[i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=10, color='#333333')

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar la escala del eje y para mayor legibilidad
plt.ylim(0, max(porcentaje_ajustado) + 5)

# Mostrar el gráfico
plt.tight_layout()
plt.show()

```
